# Data Extraction

```python
import pandas as pd
import numpy as np
import yaml
from yaml import safe_load
from tqdm import tqdm
import warnings
import os
import pickle
```

```python
files_name = []
path = r'C:\Users\Manadip\OneDrive\Desktop\ML Project\t20s'

 # Get the full File Path
for file in os.listdir(path):
    full_path = os.path.join(path , file)

# add file path to the list 
    files_name.append(full_path)
```

```python
final_df = pd.DataFrame()
counter = 1
for i in tqdm(files_name):
    with open(i , 'r') as f:
        data = yaml.safe_load(f)
        if isinstance(data , list) or isinstance(data,dict):
            df = pd.json_normalize(data)

        else:
            raise ValueError(f"Filr{i} does not contain list or dict")
        
        df['match_id'] = counter
        final_df = pd.concat([final_df , df], ignore_index  = True)
        counter += 1
        
        
```

```python
for i, row in enumerate(final_df):
    if i == 0 :
        print(row)
```

```python
for index , row in final_df.iterrows():
    if index == 700:
        delivery = row['innings'][0]['1st innings']['deliveries']
        for ball in delivery:
            ball_number = list(ball.keys())[0]
            #print(ball_number)
            ball_info  = ball[ball_number]
            if 'wicket' in ball_info:
                print(ball_info)
            #print(ball_info['wicket'])
        #print(row['innings'][0]['1st innings']['team'])
```

```python
delivery_list = []
match_id = 1
for index , row in final_df.iterrows():
    try:
        deliveries = row['innings'][0]['1st innings']['deliveries']
        batting_team_name = row['innings'][0]['1st innings']['team']
    except:
        continue

    ball_list = []
    batsman_list = []
    bolower_list  = []
    runs_list = []
    dismissed_list = []
    teams_list = []
    batting_team_list = []
    match_id_list = []
    city_list = []
    venue_list = []

    for ball in deliveries:
        ball_number = list(ball.keys())[0]
        ball_info  = ball[ball_number]

        ball_list.append(ball_number)
        batsman_list.append(ball_info.get('batsman'))
        bolower_list.append(ball_info.get('bowler'))
        runs_list.append(ball_info['runs']['total'])
        teams_list.append(row.get('info.teams'))
        batting_team_list.append(batting_team_name)
        match_id_list.append(match_id)
        city_list.append(row.get('info.city'))
        venue_list.append(row.get('info.venue'))
        if 'wicket' in ball_info:
            dismissed_list.append(ball_info["wicket"].get('player_out'))

        else:
            dismissed_list.append(0)

    # Crteating data frame for each match
    match_df = pd.DataFrame({
        'ball' : ball_list,
        'batsman': batsman_list,
        'bolower':bolower_list,
        'runs':runs_list,
        'teams':teams_list,
        'batting_team':batting_team_list,
        'match_id': match_id_list,
        'city' : city_list,
        'venue': venue_list,
        'player_dismissed': dismissed_list


    })

    delivery_list.append(match_df)
    match_id += 1

delivery_df = pd.concat(delivery_list, ignore_index=True)


```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```
