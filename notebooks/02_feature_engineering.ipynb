# Current Score

```python
print(df['runs'].dtype)
```

```python
df['current_score']= df.groupby('match_id')['runs'].cumsum()
```

```python
df.head(5)
```

```python
# wicket left , balls left
```

# Balls left

```python
df[["over","balls"]]=df['ball'].astype(str).str.split('.').apply(pd.Series)
```

```python
df.tail()
```

```python
df['ball_bowledd'] = (df['over'].astype('int')*6 ) + (df['balls'].astype('int'))
```

```python
# df = df.drop(columns='ball_bowled' , axis = 1)
```

```python
df.head(126)
```

```python
print(df['ball_bowledd'].dtype)
```

```python
df['ball_left'] = 120 - df['ball_bowledd']
df['ball_left'] = df['ball_left'].apply(lambda x: 0 if x<0 else x)
```

```python
df
```

# Wickets Left

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x==0 else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
print(df['player_dismissed'].dtype)
```

```python
df
```

# Current runrate

```python
df['crr'] = (df['current_score'] * 6)/df['ball_bowledd']
```

```python
df
```

```python
groups = df.groupby('match_id')
match_id = df['match_id'].unique()
last_five = []
for id in match_id:
    match_df = groups.get_group(id)
    last_five.extend(
        match_df['runs'].rolling(window=30).sum().values.tolist()
    )
```

```python
df['last_five'] = last_five
```

# Calculating Total Score 

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df , on = 'match_id')
```
